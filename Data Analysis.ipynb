## Exploratory Data Analysis

```python
df.columns
```

## Gender

```python
gen=sns.countplot(data=df,x='Gender')
for bar in gen.containers:
    gen.bar_label(bar)
```

```python
sale_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sale_gen)
```

## Age

```python
age=sns.countplot(data=df,x="Age Group" ,hue="Gender")
for bar in age.containers:
    age.bar_label(bar)
```

From above the graph we see that that most of the buyer are of age group between 26-35

## States

```python
state_sales=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={"figure.figsize":(21,8)})
sns.barplot(data=state_sales,x='State',y='Orders')
```

```python
state_sales=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={"figure.figsize":(21,8)})
sns.barplot(data=state_sales,x='State',y='Amount')
```

From above the graph we can see that most order are form utter pardesh,Mahrashtra amd karantaka respectively
